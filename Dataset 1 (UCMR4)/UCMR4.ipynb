```python
#merging two datasets 

# Merge datasets based on 'Contaminant' column
#ucmr3_m= pd.merge(ucmr3, ucmr3_name, on='Contaminant', how='left')

# Fill missing values in 'Method Name' column with corresponding method names
#ucmr3_m['Method Name'] = ucmr3_m['Method Name'].fillna(method='ffill')

ucmr4_m = pd.merge(ucmr4, ucmr4_name[["Contaminant","Contaminant Classification"]], on = "Contaminant", how = "left")
```

```python
#filtering data
ucmr4_m = ucmr4_m.loc[ucmr4_m['PWSName'].isin(['Henry County Water & Sewer Authority', 'DeKalb County', 'Clayton County Water Authority'])]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming ucmr4_m is your DataFrame containing the data

# Replace zeros with NaN in 'AnalyticalResultValue' column
ucmr4_m['AnalyticalResultValue'] = ucmr4_m['AnalyticalResultValue'].replace(0, np.nan)

# Convert 'CollectionDate' column to datetime format
ucmr4_m['CollectionDate'] = pd.to_datetime(ucmr4_m['CollectionDate'])

# Filter data where 'Contaminant Classification' is 'Pesticide'
disinfection_data = ucmr4_m[ucmr4_m['Contaminant Classification'] == 'Disinfection Byproducts']

# Group pesticide data by 'CollectionDate' and 'PWSName', calculate average 'AnalyticalResultValue'
grouped_df = disinfection_data.groupby(['CollectionDate', 'PWSName'])['AnalyticalResultValue'].mean().reset_index()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through each PWSName and plot with a different color
for pws_name, color in zip(grouped_df['PWSName'].unique(), ['blue', 'green', 'red']):
    pws_data = grouped_df[grouped_df['PWSName'] == pws_name]
    ax.plot(pws_data['CollectionDate'], pws_data['AnalyticalResultValue'], marker='o', linestyle='-', color=color, label=pws_name)

# Set labels and title
ax.set_xlabel('Collection Date')
ax.set_ylabel('Average Analytical Result Value (µg/L)')
ax.set_title('Average Analytical Result Value (Disinfection Byproducts) by Collection Date and County')

custom_labels = {'Henry': 'Henry', 'Clayton': 'Clayton', 'DeKalb': 'Dekalb'}  # Customize labels as per your requirement

# Show plot
plt.legend(title='County', labels=custom_labels)
# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plotz
plt.tight_layout()
plt.show()
```

```python
# Filter data where 'Contaminant Classification' is 'Pesticide'
metal_data = ucmr4_m[ucmr4_m['Contaminant Classification'] == 'Metal']

# Group pesticide data by 'CollectionDate' and 'PWSName', calculate average 'AnalyticalResultValue'
grouped_df = metal_data.groupby(['CollectionDate', 'PWSName'])['AnalyticalResultValue'].mean().reset_index()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through each PWSName and plot with a different color
for pws_name, color in zip(grouped_df['PWSName'].unique(), ['blue', 'green', 'red','yellow']):
    pws_data = grouped_df[grouped_df['PWSName'] == pws_name]
    ax.plot(pws_data['CollectionDate'], pws_data['AnalyticalResultValue'], marker='o', linestyle='-', color=color, label=pws_name)

# Set labels and title
ax.set_xlabel('Collection Date')
ax.set_ylabel('Average Analytical Result Value (µg/L)')
ax.set_title('Average Analytical Result Value (Metal) by Collection Date and County')

custom_labels = {'Henry': 'Henry', 'Clayton': 'Clayton', 'DeKalb': 'Dekalb'}  # Customize labels as per your requirement

# Show plot
plt.legend(title='County', labels=custom_labels)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plotz
plt.tight_layout()
plt.show()
```
